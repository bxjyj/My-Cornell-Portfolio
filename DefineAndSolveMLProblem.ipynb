## Part 5: Implement Your Project Plan

<b>Task:</b> In the code cell below, import additional packages that you have used in this course that you will need to implement your project plan.

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
features = [ # creating the feature list and removing redundant features from dataset
    'Life Ladder', 'Log GDP per capita', 'Social support',
    'Healthy life expectancy at birth', 'Freedom to make life choices',
    'Generosity', 'Perceptions of corruption', 'Positive affect',
    'Negative affect', 'Confidence in national government',
    'GINI index (World Bank estimate)', 'Delivery Quality'
]
X = df[features]
y = df['Democratic Quality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
# creating the training and testing split
```

```python
# creating the linear regression model 
linreg = LinearRegression()
linreg.fit(X_train, y_train)
y_pred_lin = linreg.predict(X_test)

print("Linear Regression Performance:")
print("R2:", r2_score(y_test, y_pred_lin))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_lin)))
print("MAE:", mean_absolute_error(y_test, y_pred_lin))
```

```python
# creating the Random Forest Regressor, utilizing the best parameters found by the grid search ( next cell )
rf = RandomForestRegressor(n_estimators=100, max_depth=20, random_state=1234)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

print("Random Forest Performance:")
print("R2:", r2_score(y_test, y_pred_rf))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_rf)))
print("MAE:", mean_absolute_error(y_test, y_pred_rf))
```

```python
# utilizing a grid search to tune parameters for Random Forest best performance
param_grid_rf = {
    "n_estimators": [50, 100, 200],
    "max_depth": [3, 5, 10, 20]
}

rf_cv = GridSearchCV(RandomForestRegressor(random_state=1234), param_grid_rf, cv=3, scoring="r2")
rf_cv.fit(X_train, y_train)

print("Best Random Forest Params:", rf_cv.best_params_)
print("Best CV R2:", rf_cv.best_score_)
```

```python
# comparing the performance results for linear regression and random forest
plt.scatter(y_test, y_pred_lin, alpha=0.5, label="Linear Regression")
plt.scatter(y_test, y_pred_rf, alpha=0.5, label="Random Forest")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel("True Democratic Quality")
plt.ylabel("Predicted Democratic Quality")
plt.title("Predictions vs. True Values")
plt.legend()
plt.show()
```

Above, we can see that the linear regression model has more points that are significantly farther (more higher) from the red line, which means it does not have a good fit compared to the random forest that is closer to the red line. This signifies that the performance is worse compared to the RF and is experiencing less accuracy. Let us compare performances with cross validation below.

```python
from sklearn.model_selection import cross_val_score

cv_scores_lin = cross_val_score(linreg, X_train, y_train, cv=5, scoring="r2")
cv_scores_rf = cross_val_score(rf, X_train, y_train, cv=5, scoring="r2")

cv_df = pd.DataFrame({
    "Model": ["Linear Regression", "Random Forest"],
    "Mean R2": [np.mean(cv_scores_lin), np.mean(cv_scores_rf)],
    "Std R2": [np.std(cv_scores_lin), np.std(cv_scores_rf)]
})

print(cv_df)
```

After running the models and optimizing for performance, we see that the Random Forest model performs better than the linear regression. When testing via the cross validation scores, we see that the RF has a mean R2 of 90%, which means the model accounts for 90% of the variance of the data. The linear regression model only accounts for 82% of the dataset, signifying that it has less performance than the RF. Futhermore, the standard deviation of the RF is 2% which is less than then 5% of the linear regression model. This shows that the performance is more stable in the RF when tested in the cross validations, and is more likely to generalize unseen data better.

The RF is the better model to predict the democratic quality of countries from the given dataset.
